```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
```

```python
print(data)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['Target_Price'] = data.target
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
correlation=df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
X=df.drop(['Target_Price'],axis=1)
y=df['Target_Price']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=XGBRegressor()
model.fit(X_train,y_train)
```

```python
training_data_pred=model.predict(X_train)
print(training_data_pred)
```

```python
score_R2=metrics.r2_score(training_data_pred,y_train)
print(score_R2)
```

```python
score_MAE=metrics.mean_absolute_error(y_train,training_data_pred)
print(score_MAE)
```

```python
testing_data_pred=model.predict(X_test)
print(testing_data_pred)
```

```python
score_R2=metrics.r2_score(testing_data_pred,y_test)
print(score_R2)
```

```python
score_MAE=metrics.mean_absolute_error(testing_data_pred,y_test)
print(score_MAE)
```

```python
plt.scatter(y_train,training_data_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
```
